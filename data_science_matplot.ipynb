```python
t= np.arange(0.0, 5.0, 0.2) #between 0 to 5
plt.plot(t, t, "r--", t, t**2, "bs", t, t**3, "g^", t, t**4, "ro", markersize=5) #multiple graphs
plt.show()
```

```python
veri={ "a": np.arange(50),                         # a: creates 50 numbers betwen 0 to 50
       "c": np.random.randint(0,50,50),            #c: between 0 to 50, 50 integers
       "d": np.random.randn(50) }                  #d: created number's standard deviation is 50
veri["b"] = veri["a"] + (10 * np.random.rand(50))  #b= (10*noise)+a
veri["d"] = np.abs(veri["d"])*100                  #d= absoluate value of veri * 100
plt.scatter("a","b", c="c", s="d", data=veri)
plt.xlabel("a stünundaki veriler")
plt.ylabel("b stünundaki veriler")
plt.show()
```

Multiple graphs in the same line

```python
isimler= ["Kıvılcım", "Zeynep",  " Betül", "İrem"]
degerler= [5, 25, 50, 100]
plt.figure(figsize=(9,3)) #width and height of the graphs

plt.subplot(131) #131; 1st graph
plt.bar(isimler, degerler, color="pink")

plt.subplot(132)
plt.scatter(isimler, degerler)

plt.subplot(133)
plt.plot(isimler, degerler, "b")

plt.suptitle("Farklı Grafikler") #super title
plt.show()
```

```python
mu1= 100 #ortalamsı mu1
sigma1= 15
x1= mu1 + sigma1 * np.random.randn(10000)

mu2= 90
sigma2= 20
x2= mu2 + sigma2 * np.random.randn(10000)
#histogram of the data:
n, bins, patches = plt.hist(x1, 100, density=1, facecolor="b", alpha=0.25) #alpha transparanlığı
n, bins, patches = plt.hist(x2, 100, density=1, facecolor="g", alpha=0.25)

plt.xlabel("Veriler")
plt.ylabel("Olasılıklar")
plt.title(r"Verilerin Histogramı: $\mu_1,\  \mu_2=110$") #math text, special texts should be expected after "r" $\mu is a special character
plt.annotate("Gauss eğrisinin tepesi", xy=(100, 0.03), xytext=(130, 0.045),
             arrowprops=dict(facecolor="black", shrink=0.01)) #we add an arrow, ery important
plt.text(50, 0.02, r" $\mu_1=100,\ \sigma_1=15$") #sub indices
plt.text(120, 0.03, r" $\mu_2=110,\ \sigma_2=20$")
plt.axis([40, 160, 0, 0.05])
plt.grid(True)
plt.show()
```
