```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

import warnings
warnings.filterwarnings("ignore")
```

```python
#insert the dataset
path= "/content/drive/MyDrive/BTK/DATASET/olimpiyatlar_temizlenmis_20220204.csv"
veri = pd.read_csv(path) #cvs=coma seperated value
veri.head(5)
```

```python
def sacilimGrafikCiz():
  sns.scatterplot(x="boy", y="kilo", data=veri)
  plt.title("Boy ve Ağırlık ilişkisi- Beyaz Izgara Tema")
  plt.show()
sacilimGrafikCiz()
```

```python
sns.set_style("darkgrid")
sacilimGrafikCiz()
```

İki boyutlu histogram

```python
sns.displot(veri, x="boy", y="kilo", kind="kde")
plt.xlabel("Boy")
plt.ylabel("Ağırlık")
plt.title("Boy ve Kilo Histogramı")
plt.show()
```

```python
sns.displot(veri, x="boy", y="kilo", hue="cinsiyet", kind="kde") # kernel density estimate (KDE) = yoğunluk haritaları
plt.xlabel("Boy")
plt.ylabel("Ağırlık")
plt.title("Cinsiyete Göre Ayrıştırılmış Boy ve Kilo Histogramı")
plt.show()
```

```python
sns.kdeplot(data=veri, x="boy", y="kilo", hue="cinsiyet",fill="True")
plt.xlabel("Boy")
plt.ylabel("Ağırlık")
plt.title("Cinsiyete Göre Ayrıştrılmoş Boy-Kilo Histogramı")
plt.show()
```

```python
g=sns.JointGrid(y="boy", x="kilo", data=veri) #buradaki syntax farklı
g.plot_joint(sns.histplot)
g.plot_marginals(sns.boxplot) #kenarda istenen grafik verildi
plt.show()
```

Pairplots

```python
sns.pairplot(veri)   #burada çizilebilecek tüm grafikleri görmek için bu konu yazdım
plt.show()           #şekilden görüldüğü üzere 4x4'lük grafik çizildi
```

```python
g= sns.PairGrid(veri)
g.map_upper(sns.histplot)
g.map_lower(sns.kdeplot, fill=True)
g.map_diag(sns.histplot, kde=True)
plt.show()
```

```python
sns.countplot(x="sehir", data=veri, palette="Set2") #her şehirde kaç madalya verildi, kaç kez olimpiyat yapıldı?Bunu görmek istiyorum
plt.xticks(rotation=90)
plt.show()
```
